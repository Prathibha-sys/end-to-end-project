```python
import pandas as pd
df=pd.read_csv("/content/data.csv")
df.head()
```

```python
df=df.drop_duplicates()
```

```python
#drop date column
df=df.drop(columns='date')
```

```python
df.head()
```

```python
#Find house age
df['House_Age']=2026-df['yr_renovated']-df['yr_built']

df.head()
```

```python
df=df[df['House_Age']>0]
```

```python
df.head()
```

```python
#EDA Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Understand Price Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['price'],kde=True,bins=50,palette='paired')
plt.title("Distribution of House Prices",fontsize=16,fontweight='bold')
plt.xlabel("Price",fontsize=12)
plt.ylabel("Count",fontsize=12)
plt.show()
```

```python
#Area Vs Price(Core Relationship)
plt.figure(figsize=(8,5))
sns.scatterplot(df,x='sqft_living',y='price',color='red',hue='House_Age')
plt.title("Area Vs Price",fontsize=16,fontweight='bold')
plt.xlabel("Area",fontsize=12)
plt.ylabel("Price",fontsize=12)
plt.show()
```

```python
#House Age Vs Price
plt.figure(figsize=(8,5))
sns.lineplot(df,x='House_Age',y='price')
plt.title("House Age Vs Price",fontsize=16,fontweight='bold')
plt.xlabel("House Age",fontsize=12)
plt.ylabel("Price",fontsize=12)
plt.show()
```

```python
#Floors VS Price
plt.figure(figsize=(8,5))
sns.barplot(df,x='floors',y='price',palette='rocket')
plt.title("Floors Vs Price",fontsize=16,fontweight='bold')
plt.xlabel("Floors",fontsize=12)
plt.ylabel("Price",fontsize=12)
plt.show()
```

```python
#Bedroom Vs Price(Boxplot)
plt.figure(figsize=(8,5))
sns.boxplot(df,x='bedrooms',y='price',palette='rocket')
plt.title("Bedroom Vs Price",fontsize=16,fontweight='bold')
plt.xlabel("Bedrooms",fontsize=12)
plt.ylabel("Price",fontsize=12)
plt.show()
```

```python
#Bedrooms Vs Average price
bedroom_avg=df.groupby('bedrooms')['price'].mean().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(bedroom_avg,x='bedrooms',y='price',palette='rocket')
plt.title("Bedrooms Vs Average Price",fontsize=16,fontweight='bold')
plt.xlabel("Bedrooms",fontsize=12)
plt.ylabel("Average Price",fontsize=12)
plt.show()
```

```python
#Correlation Heatmap
num  = df.select_dtypes(include = 'number')
plt.figure(figsize=(10,8))
sns.heatmap(num.corr(),annot=True)
plt.title("Correlation Heatmap",fontsize=16,fontweight='bold')
plt.show()
```

```python
#Price Per Sqft
df['price_per_sqft']=df['price']/df['sqft_living']
sns.histplot(df['price_per_sqft'],kde=True,bins=50,palette='rocket')
plt.title("Distribution of Price Per Sqft",fontsize=16,fontweight='bold')
plt.xlabel("Price Per Sqft",fontsize=12)
plt.ylabel("Count",fontsize=12)
plt.show()
```

```python
#Top 10 expensive Houses
top10=df.sort_values(by='price',ascending=False).head(10)
sns.barplot(top10,x='price',y=top10.index,palette='rocket')
plt.title("Top 10 Expensive Houses",fontsize=16,fontweight='bold')
plt.xlabel("Price",fontsize=12)
plt.ylabel("House ID",fontsize=12)
plt.show()

#
```
